Machine learning models oparate with numerical data therfore any column contain categorical values must me transformed into numeric format. The Gender column has categorical values such has male and female.

One of the method we can utilize to transform categorical values into numerical format is through **One Hot Encoding.**
One Hot Encoding is a way to convert catogorical data into a format that can be understood by machine learning algorithms. It creats a binary vector for each category. In a categorical variable where only one element vector is set to 1 and all other elements set to 0. This technique helps the algorithm to identify and differentiate between different categories without assuming any order or hierarchy among them.
Pandas has get a dummies method. That allows us to perform One Hot Encoding that transform categorical values into numerical values.Lets apply this method to Gender Column.

**Feature Engineering**

```python
gender_mapping = {'Female': 0, 'Male': 1}
```

```python
df['Gender_Numeric'] = df['Gender'].map(gender_mapping)
```

```python
print(df[['Gender', 'Gender_Numeric']])
```

```python
df
```

```python
# features or indipendent variables
x = df[['Gender_Numeric','Age','Salary','Price']].to_numpy()
```

```python
x
```

```python
# dependent variable
y = df['Purchased'].to_numpy()
```

```python
y
```

Splitting data into Training set and Test set. The traing set is used to train the machine learning model and test set to evaluate the performance of the trained model on new unseen data. We use train_test_split function from Scikit learn to split data set into 80 - 20 ratio. The 80% of the data is used for traing and 20% for testing. The exact splitting ratio is dependent on the size of the data set and the cpmplexity of a model and specific requiredments of the problem is being addressed.

```python
from sklearn.model_selection import train_test_split
```

```python
# To ensure ability of our results setting random_state. This way we run same code
# with same random_state will be at the same splitting of data everytime.
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state = 42)
```

**Scaling the Features**

Scaling the features machine learning is important because many algorithms used to distance to measures the similarity between different features. If features are not scaled properly features with features with langer numeric ranges condominate the distance measure making difficult for the algorithm to learn from other features.

Scaling helps to ensure that all the features are equally weighted and the algorithm learn all the features eqully.

```python
from sklearn.preprocessing import StandardScaler
```

**Standard Scaler** standardizes features by removing the mean and scaling to unit variance. It transforms the data sets into mean of Zero(0) and standard deviation of one(1). It contributes all the featues eqully into the model it prevents featues into larger values from dominating the model.

```python
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

**Traing The Model**

**KNN Algorithm**
The k-nearest neighbors (KNN) algorithm is a non-parametric, supervised learning method used for both classification and regression tasks.

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k= 5
model = KNeighborsClassifier(n_neighbors=k)
```

```python
model.fit(x_train,y_train)
```

**Testing the Data**

Testing test data set. Use the model to predict the outputs for test set and compare predicted outputs vales with actual values to measure the accuracy of the model.

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```
